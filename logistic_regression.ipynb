```python
from sklearn.datasets import load_iris
```

```python
dt=load_iris()
```

```python
dt.keys()
```

```python
import pandas as pd
import numpy as np
```

```python
df=pd.DataFrame(dt.data,columns=dt.feature_names)
```

```python
df.info()
```

```python
df['target']=dt.target
```

```python
df.tail(10)
```

```python
df['target'].value_counts()
```

```python
df_copy=df[df['target']!=2]
```

```python
x=df_copy.iloc[:,:-1]
y=df_copy.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
classfier=LogisticRegression()
classfier
```

```python
classfier.fit(x_train,y_train)
```

```python
y_pred=classfier.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
print(confusion_matrix(y_test,y_pred))
print(accuracy_score(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

```python
from sklearn.model_selection import KFold,cross_val_score
cv=KFold(n_splits=5)
```

```python
cv
```

```python
cross_val_score(classfier,x_train,y_train,cv=cv,scoring='accuracy')
```

# multiclass classification

```python
from sklearn.datasets import make_classification
from sklearn.linear_model import LogisticRegression

x,y=make_classification(n_samples=1000, n_features=10, n_informative=5, n_redundant=5, n_classes=3, random_state=1)
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression(multi_class='ovr',solver='lbfgs')
```

```python
model
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=42)
```

```python
model.fit(x_train,y_train)
```

```python
model.predict(x_test)
```

```python
y_test
```

```python
model.predict_proba(x_test)
```

```python
y_pred_12=model.predict(x_test)
```

```python
print(confusion_matrix(y_test,y_pred_12))
print(accuracy_score(y_test,y_pred_12))
print(classification_report(y_test,y_pred_12))
```
